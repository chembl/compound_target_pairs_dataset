```python
import pandas as pd
import numpy as np
from rdkit import DataStructs
from rdkit import Chem
from rdkit import DataStructs
from rdkit.Chem import Descriptors
from rdkit.Chem import PandasTools
from rdkit.Chem import AllChem
from rdkit.Chem import Draw

data = df2
PandasTools.AddMoleculeColumnToFrame(data,'canonical_smiles','mol',includeFingerprints=False)

data['mol'] = data.canonical_smiles.apply(Chem.MolFromSmiles)

data.loc[:,'FractionCSP3'] = data['mol'].apply(Descriptors.FractionCSP3)
data.loc[:,'NumAliphaticCarbocycles'] = data['mol'].apply(Descriptors.NumAliphaticCarbocycles)
data.loc[:,'NumAliphaticHeterocycles'] = data['mol'].apply(Descriptors.NumAliphaticHeterocycles)
data.loc[:,'NumAliphaticRings'] = data['mol'].apply(Descriptors.NumAliphaticRings)
data.loc[:,'NumAromaticCarbocycles'] = data['mol'].apply(Descriptors.NumAromaticCarbocycles)
data.loc[:,'NumAromaticHeterocycles'] = data['mol'].apply(Descriptors.NumAromaticHeterocycles)
data.loc[:,'NumAromaticRings'] = data['mol'].apply(Descriptors.NumAromaticRings)
data.loc[:,'NumHeteroatoms'] = data['mol'].apply(Descriptors.NumHeteroatoms)
data.loc[:,'NumSaturatedCarbocycles'] = data['mol'].apply(Descriptors.NumSaturatedCarbocycles)
data.loc[:,'NumSaturatedHeterocycles'] = data['mol'].apply(Descriptors.NumSaturatedHeterocycles)
data.loc[:,'NumSaturatedRings'] = data['mol'].apply(Descriptors.NumSaturatedRings)
data.loc[:,'RingCount'] = data['mol'].apply(Descriptors.RingCount)
```
